Merging two tables

```python
final_df= pd.merge(df,df_country ,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
country_values=final_df.Country.value_counts().values
```

```python
country_names=final_df.Country.value_counts().index
```

Pie chart for volume country wise

```python
plt.pie(country_values,labels=country_names)
```

```python
plt.pie(country_values[:3],labels=country_names[:3])
```

pie chart for volume percentage wise per country

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation till now: Companies most business is done in India apart from this US and then UK are major marjets.

Grouping 3 columns which are related to rating in one

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index(name ='RatingCount')
```

```python
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='RatingCount',data=ratings)
```

```python
plt.rcParams['figure.figsize']=(12,8)
sns.barplot(x='Aggregate rating',y='RatingCount' ,hue='Rating color',data=ratings)
```

```python
plt.rcParams['figure.figsize']=(12,8)
sns.barplot(x='Aggregate rating',y='RatingCount',hue='Rating color',data=ratings, palette=['white','red','yellow','orange','green','blue'])
```

Observation:
1. Not rated count is very high.
2. MAx ratings are between 2.5 to 4.1


```python
sns.countplot(x='Rating color',data=ratings, palette=['white','red','yellow','orange','green','blue'])
```
